## Derivitative Concepts
Defined in two scopes:
- **Geometric**
  - The derivative is the slope of a curve. 
  - A tangent line at any line of the curve, that line would be the derivative of the curve. 
  - Taking the derivative of the curve gives us the slope of that curve. 
  - Essentialy, this lets us know how much this curve is changing at this particular point.
- **Physical**
  - The derivative is the rate of change of any given function
  - If we have an equation describing the displacement of an object, if we take the derivative we would get the velocity.
  - Velocity describes how quickly the object is physically chaning
  - Then, we can take the derivative of the velocity, and we get the acceleration. Acceleration is the rate of change of velocity
  - We understand how fast, and in what direction the rate is changing

## Derivatives - The Basics
- The derivative operator is written as
$\frac{d}{dx}$
- $d$ is the derivative
- Bottom $x$ tells us what variable we are taking the derivative with respect to
- We write derivatives as
  - If $y = f(x)$ then $\frac{dy}{dx}$ = $\frac{df(x)}{d(x)}$ = $\frac{d}{dx}f(x)$
  - $f\prime(x)=\frac{d}{dx}f(x)$

## Scalar Derivative Rules

### Constant
- The derivative of any constant is 0
- $\frac{d}{dx}c=0$
- Because there is no $x$ in the function
- $x$ is not changing, so it makes sense that it evaluates to 0
- Example
  - $\frac{d}{dx}38=0$

### Power Rule
- The most important amd most commonly used
- Handles variables in derivaties
- $\frac{d}{dx}x^{n}=nx^{n-1}$
- Example
  - $\frac{d}{dx}x^{3}=3x^{2}$
  
### Multiplication
- $\frac{d}{dx}cx^{n}=c\frac{d}{dx}x^{n}=cnx^{n-1}$
- Example
  - $\frac{d}{dx}4x^{3}=4\frac{d}{dx}x^{3}=4*3x^{2}=12x^{2}$
  
### The Sum Rule
- $\frac{d}{dx}(f(x)+g(x)) = \frac{d}{dx}f(x) + \frac{d}{dx}g(x)$
- Example
  - $\frac{d}{dx}(4x+2x^{2})=4\frac{d}{dx}x^{1}+2\frac{d}{dx}x^{2}=4+4x$
  
### The Product Rule
- $\frac{d}{dx}(f(x)g(x)) = g(x)\frac{d}{dx}f(x) + f(x)\frac{d}{dx}g(x)$
- Example
  - $\frac{d}{dx}(x^{2}x) = x\frac{d}{dx}x^{2} + x^{2}\frac{d}{dx}x = x(2x) + x^{2} = 2x^{2}+x^{2}=3x^{2}$
  
### The Chain Rule
- Helpful in calculating the derivatives of layered systems
- $\frac{d}{dx}(f(g(x)) = f\prime(g(x))y\prime(x))$
- Example
  - Note: THe derivative of $\sin$ is $\cos$
  - $\frac{d}{dx}sin(x^{2}) = \cos(x^{2})*2x=2x\cos(x^2)$

## Introduction to Partials
- Partial derivatives are used for equations with more than one variable
- The most common in Machine Learning
- $\frac{d}{dx}$ -> Partial with respect to $x$
- $\frac{d}{dy}$ -> Partial with respect to $y$
- Example: $f(x,y) = 3x^{2}y$
  - $\frac{d}{dx}f(x,y) = \frac{d}{dx}3x^{2}y=3y\frac{d}{dx}x^{2}=3y(2x)=6xy$
    - Note: Taking the partial with respect to x, we treat y as constant, so y becomes now as a coefficient
  - $\frac{d}{dy}f(x,y) = \frac{d}{dy}3x^{2}y=3x^{2}\frac{d}{dy}y=3x^{2}(1)=3x^{2}$

## Basics of Integration
- Related to Probability and Probability Theory
- Integration is used to find the area under a curve
- Integration is the reverse of differentiation (derivatives); as a result, the **integral** is often called the **anti-derivative**
- ![](imgs/Lecture-03/derint.gif)
- Integral finds the area under the curve
- If the derivative of a space equation told us the velocity, if we take the integral of the velocity we could figure out how far out that object went in that time

## Types of Integrals
- Indefinite
  - Integrals with no limits
  - from $-\infty$ to $\infty$
  - $\int f(x)dx$
  
- Definite 
  - Integrals with limits
  - $\int_{a}^{b}f(x)d(x)$

### The Anti-derivative
- Integration reverses differentiation
- $\int f\prime(x)dx = f(x) + C$
  - C
    - Constant of Integration
    - Remember: We have no way of knowing whether there were a constant or not, because derivative of C is always 0

## The Rules of Integration

## The Power Rule
- $\int x^{n}dx = \frac{x^{n+1}}{n+1} + C $
- Example
  - $\int x^{2}dx = \frac{x^{3}}{3} + C $
  
## Constants
- $\int kdx = kx + C $
- Example
  - $\int 4dx = 4x + C $

## Evaluating Definite Integrals
$$\int_{a}^{a}f\prime(x)dx = f(x) + C$$
- (value of f(x) + C at x=b) - (value of f(x) + C at x=a)
$$\int_{a}^{b}f\prime(x)dx=f(b)-f(a)$$
$$\int_{a}^{b}f\prime(x)dx = f(x)\vert_{a}^{b}=f(b)-f(a)$$

## Vector Calculus: Gradients
- Partial derivatives can be organized into a vector
- $\frac{d}{dx} and \frac{d}{dy}$
- can be organized to be
  - $\nabla = [\frac{d}{dx}, \frac{d}{dy}]$
  - This operator $\nabla$ is called **nabla** operator
  - It refers to the **Gradient**
- $\nabla f(x,y)=[\frac{df(x,y)}{dx},\frac{df(x,y)}{dy}]$
- Gradient example in two dimensions
  - $f(x,y) = 3x^{2}y$
  - $\nabla 3x^{2}y = [\frac{d}{dx}3x^{2}y, \frac{d}{dy}3x^{2}y]=[6xy, 3x^{2}]$

## Understanding the Gradient Conceptually
- Like the derivative, the gradient represents the slope of a function
- The gradient points in the direction of the greatest rate of increase of the function, and its magnitude is the slope in that direction
- Gradient Vectors
  - Gradients are not restricted to two diemnsions, they can easily be expanded
  - $\nabla f(x,y,z) = [\frac{df(x,y,z)}{dx}, \frac{df(x,y,z)}{dy}, \frac{df(x,y,z)}{dz}]$
  - The **Jacobian** Matrix
    - Gradient vectors organize the partial derivatives for a scalar function. If we have multiple functions, we use the **Jacobian**
    - $J = \begin{bmatrix}\nabla \int(x,y) \\ \nabla g(x,y) \end{bmatrix} = \begin{bmatrix} \frac{df(x,y)}{dx} & \frac{df(x,y)}{dy} \\ \frac{dg(x,y)}{dx} & \frac{dg(x,y)}{dy} \end{bmatrix}$
  - The **General** Case
$$f(x,y,z) \rightarrow f(x) where x = \begin{bmatrix}x1 \\ x2 \\ . \\ . \\ . \\ xn \end{bmatrix}$$
$$J = \begin{bmatrix}\nabla f_{1}(x) \\ \nabla f_{2}(x) \\ ... \\ \nabla f_{m}(x) \end{bmatrix} = \begin{bmatrix} \frac{d}{dx1}f_{1}(x) & \frac{d}{dx2}f_{1}(x) & ... & \frac{d}{dx_{n}}f_{1}(x)\\ ... & ... \end{bmatrix}$$

### Gradient Visualization with Matplotlib

```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
print('Python version: {}'.format(sys.version))
print('Numpy version: {}'.format(np.__version__))
print('Matplotlib version: {}'.format(matplotlib.__version__))
```

```python
# generate 2D meshgrid
nx, ny = (100,100)

x = np.linspace(0,10, nx)
y = np.linspace(0,10, ny)

xv, yv = np.meshgrid(x,y)
print(xv)
```

```python
print(xv.shape)
```

```python
# define a function to plot
def f(x,y):
    return x * (y** 2)
```

```python
# calculate Z value for each x,y point
z = f(xv, yv)
print(z)
```

```python
print(z.shape)
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title('2D Color Plot of f(x,y)=xy^2')
plt.colorbar()
plt.show()
```

```python
# calculate gradients using numpy
# generate new 2D meshgrid
# make it smaller
nx, ny = (10,10)
x = np.linspace(0,10,nx)
y = np.linspace(0,10,ny)
xg, yg = np.meshgrid(x,y)

# calculate the gradient of f(x,y)
Gy, Gx = np.gradient(f(xg,yg))
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv, yv, z)
plt.title('Gradient of f(x,y) = xy^2')
plt.colorbar()
plt.quiver(xg, yg, Gx, Gy, scale=1000, color='w')
plt.show()
```
